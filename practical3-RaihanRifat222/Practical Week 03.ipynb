# 1. Convolution

The following code uses OpenCV's filter2D to define a convolution filter. You can read a tutorial that demonstrates the use of OpenCV's filter2D here:
  - [Tutorial filter 2D](https://docs.opencv.org/3.4/d4/dbd/tutorial_filter_2d.html)

```python
import cv2
import numpy as np
def convolution_filter(image, kernel):
    return cv2.filter2D(image, -1, kernel)
```

Below is an example of use of this filter for the first training sample of the MNIST dataset.

```python
from tensorflow.keras.datasets import mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
```

```python
kernel = np.array([[-1, -1, -1],
                   [-1, 8, -1],
                   [-1, -1, -1]])
result = convolution_filter(train_images[0], kernel)
```

```python
from matplotlib import pyplot as plt
plt.imshow(train_images[0], cmap='gray')
```

```python
plt.imshow(result, cmap='gray')
```

```python
img = plt.imread('stickanimalsGray.png')
plt.imshow(img, cmap=  'gray')
```

```python
edge_detection_kernel = np.array([[-1, -1, -1],
                   [-1, 8, -1],
                   [-1, -1, -1]])


edge_detected = convolution_filter(img, edge_detection_kernel)

plt.imshow(edge_detected, cmap=  'gray')
```

```python
sharpen_kernel = np.array([[0, -1, 0],        [-1, 5, -1],        [0, -1, 0]])


sharpen = convolution_filter(img, sharpen_kernel)

plt.imshow(sharpen, cmap=  'gray')
```

```python
blur_kernel = np.array([[1/9, 1/9, 1/9],        [1/9, 1/9, 1/9],        [1/9, 1/9, 1/9]])


blurred = convolution_filter(img, blur_kernel)

plt.imshow(blurred, cmap=  'gray')
```
